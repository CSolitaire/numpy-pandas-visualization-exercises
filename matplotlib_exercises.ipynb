```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
x = np.array(range(10))
y = ((x * x) - x + 2)
plt.plot(y)
plt.annotate('Origin', xy = (0,2), xytext = (0,20), arrowprops = {'arrowstyle':'->'})
plt.title ('$Y=X^2-X+2$')
plt.show()
```

```python
x = np.array(range(10))
y = (x ** 0.5)
plt.title ('$Y=\sqrt{2}$')
plt.plot(y)
plt.show()
```

```python
x = np.array(range(10))
y = (x ** 3)
plt.title ('$Y=X^3$')
plt.plot(y)
plt.show()
```

```python
x = np.array(range(10))
y = (2 ** x)
plt.title ('$Y=2^X$')
plt.plot(y)
plt.show()
```
