```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

## 1. Use matplotlib to plot the following equation: y=x*x−x+2
-     You'll need to write the code that generates the x and y points.
-     Add an anotation for the point 0, 0, the origin.

```python
x = range(0,10)
y = [(n * n) - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy = (0,2), xytext = (0,20), arrowprops = {'arrowstyle':'->'})
plt.title ('$Y=X^2-X+2$')
plt.show()
```

```python
x = np.array(range(10))
y = (2 ** x)
plt.title ('$Y=2^X$')
plt.plot(y)
plt.show()
```

## 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (15,8))

plt.subplot(2, 2, 1)
x = np.array(range(10))
y = ((x * x) - x + 2)
plt.plot(y)
plt.annotate('Origin', xy = (0,2), xytext = (0,20), arrowprops = {'arrowstyle':'->'})
plt.title ('$Y=X^2-X+2$')

plt.subplot(2, 2, 2)
x = np.array(range(10))
y = (x ** 0.5)
plt.title ('$Y=\sqrt{2}$')
plt.plot(y)

plt.subplot(2, 2, 3)
x = np.array(range(10))
y = (x ** 3)
plt.title ('$Y=X^3$')
plt.plot(y)

plt.subplot(2, 2, 4)
x = np.array(range(10))
y = (2 ** x)
plt.title ('$Y=2^X$')
plt.plot(y)

plt.suptitle("Four Function Graphic", fontsize = 20)
plt.tight_layout(pad = 3.0)
plt.show ()
```

## 4. Combine the figures you created in the last step into one figure where each of the 4              equations has a different color for the points. Be sure to include a legend.

```python
plt.figure(figsize = (15,8))

plt.subplot(2, 2, 1)
x = np.array(range(10))
y = ((x * x) - x + 2)
plt.plot(y, c ='purple', label = '$Y=X^2-X+2$')
plt.annotate('Origin', xy = (0,2), xytext = (0,20), arrowprops = {'arrowstyle':'->'})
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=X^2-X+2$')

plt.subplot(2, 2, 2)
x = np.array(range(10))
y = (x ** 0.5)
plt.plot(y, c ='orange', label = '$Y=\sqrt{2}$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=\sqrt{2}$')


plt.subplot(2, 2, 3)
x = np.array(range(10))
y = (x ** 3)
plt.plot(y, c = 'blue', label = '$Y=X^3$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=X^3$')


plt.subplot(2, 2, 4)
x = np.array(range(10))
y = (2 ** x)
plt.plot(y, c = 'green', label = '$Y=2^X$')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=2^X$')


plt.suptitle("Four Function Graphic", fontsize = 20)
plt.tight_layout(pad = 3.0)
plt.show
```
