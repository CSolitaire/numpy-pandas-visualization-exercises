```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(range(10))
y = ((x * x) - x + 2)

plt.plot(x,y)
plt.annotate('Origin', xy = (0,2), xytext = (0,20), arrowprops = {'arrowstyle':'->'})
plt.title ('$Y=X^2-X+2$')
plt.show()
```
