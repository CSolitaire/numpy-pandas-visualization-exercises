## 1. Use matplotlib to plot the following equation: y=x*xâˆ’x+2
-     You'll need to write the code that generates the x and y points.
-     Add an anotation for the point 0, 0, the origin.

```python
x = range(-20,20)
y = [(n * n) - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy = (0,0), xytext = (0,100), arrowprops = {'arrowstyle':'->'})
plt.title ('$Y=X^2-X+2$')
plt.show()
```

## 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize = (15,8))

plt.subplot(2, 2, 1)
x = range(-20,20)
y = [(n * n) - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy = (0,0), xytext = (0,100), arrowprops = {'arrowstyle':'->'})
plt.title ('$Y=X^2-X+2$')


plt.subplot(2, 2, 2)
x = list(range(20))
y = [m.sqrt(n) for n in x]
plt.title ('$Y=\sqrt{2}$')
plt.plot(x,y)


plt.subplot(2, 2, 3)
x = list(range(20))
y = [n ** 3 for n in x]
plt.title ('$Y=X^3$')
plt.plot(x,y)


plt.subplot(2, 2, 4)
x = list(range(20))
y = [2 ** n for n in x]
plt.title ('$Y=2^X$')
plt.plot(x,y)


plt.suptitle("Four Function Graphic", fontsize = 20)
plt.tight_layout(pad = 3.0)
plt.show ()
```

## 4. Combine the figures you created in the last step into one figure where each of the 4              equations has a different color for the points. Be sure to include a legend.

```python
plt.figure(figsize = (15,8))

plt.subplot(2, 2, 1)
x = range(-20,20)
y = [(n * n) - n + 2 for n in x]
plt.plot(x,y, c ='purple', label = '$Y=X^2-X+2$')
plt.scatter(x,y, c ='purple')
plt.annotate('Origin', xy = (0,0), xytext = (0,100), arrowprops = {'arrowstyle':'->'})
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=X^2-X+2$')


plt.subplot(2, 2, 2)
x = list(range(20))
y = [m.sqrt(n) for n in x]
plt.plot(x,y, c ='orange', label = '$Y=\sqrt{2}$')
plt.scatter(x,y, c ='orange')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=\sqrt{2}$')


plt.subplot(2, 2, 3)
x = list(range(20))
y = [n ** 3 for n in x]
plt.plot(x,y, c = 'blue', label = '$Y=X^3$')
plt.scatter(x,y, c = 'blue')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=X^3$')


plt.subplot(2, 2, 4)
x = list(range(20))
y = [2 ** n for n in x]
plt.plot(x,y, c = 'green', label = '$Y=2^X$')
plt.scatter(x,y, c = 'green')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.legend(loc ='upper left')
plt.title ('$Y=2^X$')


plt.suptitle("Four Function Graphic", fontsize = 20)
plt.tight_layout(pad = 3.0)
plt.show
```

```python
x = list(range(0, 25))

# generate y for each fucntion
y1 = [m.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]

plt.plot(x, y1, c='slateblue', label='$\sqrt{x}$', alpha=.5)
plt.plot(x, y2, c='orange', label='$x^{3}$', alpha=.5)
plt.plot(x, y3, c='deeppink', label='$2^{x}$', alpha=.5)
plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('My Functions')
plt.legend()

plt.show()
```
